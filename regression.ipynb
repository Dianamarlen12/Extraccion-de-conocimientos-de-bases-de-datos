# ¬øQu√© es la regresi√≥n?
La regresi√≥n funciona estableciendo una relaci√≥n entre las variables en los datos que representan caracter√≠sticas, conocidas como caracter√≠sticas, de lo que se observa y la variable que estamos tratando de predecir, conocida como etiqueta . Recuerde nuestra empresa que alquila bicicletas y quiere predecir el n√∫mero esperado de alquileres en un d√≠a determinado. En este caso, las caracter√≠sticas incluyen elementos como el d√≠a de la semana, el mes, etc., mientras que la etiqueta es el n√∫mero de alquileres de bicicletas.

Para entrenar el modelo, comenzamos con una muestra de datos que contiene las caracter√≠sticas, as√≠ como valores conocidos para la etiqueta, por lo que en este caso necesitamos datos hist√≥ricos que incluyan fechas, condiciones clim√°ticas y el n√∫mero de alquileres de bicicletas.

Luego, dividiremos esta muestra de datos en dos subconjuntos:

* Un conjunto de datos de entrenamiento al que aplicaremos un algoritmo que determina una funci√≥n que encapsula la relaci√≥n entre los valores de la caracter√≠stica y los valores de etiqueta conocidos.
* Un conjunto de datos de validaci√≥n o prueba que podemos usar para evaluar el modelo us√°ndolo para generar predicciones para la etiqueta y compararlas con los valores de etiqueta conocidos reales.
El uso de datos hist√≥ricos con valores de etiqueta conocidos para entrenar un modelo hace que la regresi√≥n sea un ejemplo de aprendizaje autom√°tico supervisado .

Hay varias formas en que podemos medir la varianza entre los valores predichos y reales, y podemos usar estas m√©tricas para evaluar qu√© tan bien predice el modelo.

 Nota

El aprendizaje autom√°tico se basa en estad√≠sticas y matem√°ticas, y es importante conocer los t√©rminos espec√≠ficos que usan los estad√≠sticos y matem√°ticos (y, por lo tanto, los cient√≠ficos de datos). Puede pensar en la diferencia entre un valor de etiqueta predicho y el valor de etiqueta real como una medida de error . Sin embargo, en la pr√°ctica, los valores "reales" se basan en observaciones de muestra (que a su vez pueden estar sujetas a alguna variaci√≥n aleatoria). Para dejar en claro que estamos comparando un valor predicho ( ≈∑ ) con un valor observado ( y ), nos referimos a la diferencia entre ellos como los residuos. Podemos resumir los residuos de todas las predicciones de datos de validaci√≥n para calcular la p√©rdida general en el modelo como una medida de su rendimiento predictivo.

Una de las formas m√°s comunes de medir la p√©rdida es elevar al cuadrado los residuos individuales, sumar los cuadrados y calcular la media. Cuadrar los residuos tiene el efecto de basar el c√°lculo en valores absolutos (ignorando si la diferencia es negativa o positiva) y dar m√°s peso a las diferencias m√°s grandes. Esta m√©trica se llama Error cuadr√°tico medio .

Para nuestros datos de validaci√≥n, el c√°lculo se ve as√≠:

Entonces, la p√©rdida para nuestro modelo basado en la m√©trica MSE es 9.79.

Entonces, ¬øeso es bueno? Es dif√≠cil saberlo porque el valor de MSE no se expresa en una unidad de medida significativa. Sabemos que cuanto menor es el valor, menor p√©rdida hay en el modelo; y por lo tanto, mejor es la predicci√≥n. Esto hace que sea una m√©trica √∫til para comparar dos modelos y encontrar el que funcione mejor.

A veces, es m√°s √∫til expresar la p√©rdida en la misma unidad de medida que el valor de la etiqueta predicha, en este caso, la cantidad de alquileres. Es posible hacer esto calculando la ra√≠z cuadrada del MSE, que produce una m√©trica conocida, como era de esperar, como Root Mean Squared Error (RMSE).

‚àö9,79 = 3,13

Entonces, el RMSE de nuestro modelo indica que la p√©rdida es un poco m√°s de 3, lo que puede interpretar libremente como que significa que, en promedio, las predicciones incorrectas son incorrectas en alrededor de 3 alquileres.

Hay muchas otras m√©tricas que se pueden utilizar para medir la p√©rdida en una regresi√≥n. Por ejemplo, R 2 (R-Squared) (a veces conocido como coeficiente de determinaci√≥n ) es la correlaci√≥n entre x y y al cuadrado. Esto produce un valor entre 0 y 1 que mide la cantidad de varianza que puede explicar el modelo. Por lo general, cuanto m√°s se acerca este valor a 1, mejor predice el modelo.

# Explore los datos

Comencemos cargando los datos de uso compartido de bicicletas como un Pandas DataFrame y viendo las primeras filas.

Los datos constan de las siguientes columnas:

* **instant :** un identificador de fila √∫nico
* **dteday :** la fecha en la que se observaron los datos; en este caso, los datos se recopilaron diariamente; por lo que hay una fila por fecha.

* **a√±o :** el a√±o del estudio en el que se realiz√≥ la observaci√≥n (el estudio se llev√≥ a cabo durante dos a√±os; el a√±o 0 representa 2011 y el a√±o 1 representa 2012)
* **mth :** El mes calendario en el que se realiz√≥ la observaci√≥n (1: enero ... 12: diciembre)
* **feriado :** un valor binario que indica si la observaci√≥n se realiz√≥ o no en un feriado p√∫blico)
* **weekday :** el d√≠a de la semana en el que se realiz√≥ la observaci√≥n (0: domingo ... 6: s√°bado)
* **d√≠a laborable :** un valor binario que indica si el d√≠a es un d√≠a laborable (no un fin de semana o feriado)
* **weathersit :** un valor categ√≥rico que indica la situaci√≥n meteorol√≥gica (1: despejado, 2: niebla / nube, 3: lluvia ligera / nieve, 4: lluvia intensa / granizo / nieve / niebla)

* **atemp :** La temperatura aparente ("se siente como") en grados Celsius (normalizada)
* **zumbido :** el nivel de humedad (normalizado)
* **velocidad del viento :** la velocidad del viento (normalizada)
* **alquileres :** el n√∫mero de alquileres de bicicletas registrados.
En este conjunto de datos, los **alquileres** representan la etiqueta (el valor de y ) que nuestro modelo debe estar entrenado para predecir. Las otras columnas son caracter√≠sticas potenciales ( valores x ).

Como se mencion√≥ anteriormente, puede realizar alguna ingenier√≠a de funciones para combinar o derivar nuevas funciones. Por ejemplo, agreguemos una nueva columna llamada **d√≠a** al marco de datos extrayendo el componente de d√≠a de la columna **dteday** existente .
```python
bike_data['day'] = pd.DatetimeIndex(bike_data['dteday']).day
bike_data.head(32)
```

Bien, comencemos nuestro an√°lisis de los datos examinando algunas estad√≠sticas descriptivas clave. Podemos usar el m√©todo de **descripci√≥n** del marco de datos para generarlos para las caracter√≠sticas num√©ricas, as√≠ como para la columna de etiquetas de **alquileres .**

```python
numeric_features = ['temp', 'atemp', 'hum', 'windspeed']
bike_data[numeric_features + ['rentals']].describe()
```

Las estad√≠sticas revelan alguna informaci√≥n sobre la distribuci√≥n de los datos en cada uno de los campos num√©ricos, incluido el n√∫mero de observaciones (hay 731 registros), la media, la desviaci√≥n est√°ndar, los valores m√≠nimo y m√°ximo y los valores de cuartil (los valores de umbral para 25%, 50%, que tambi√©n es la mediana y el 75% de los datos). A partir de esto, podemos ver que el n√∫mero medio de alquileres diarios es de alrededor de 848; pero hay una desviaci√≥n est√°ndar comparativamente grande, lo que indica una gran variaci√≥n en el n√∫mero de alquileres por d√≠a.

Podr√≠amos tener una idea m√°s clara de la distribuci√≥n de los valores de los alquileres visualizando los datos. Los tipos de gr√°ficos comunes para visualizar distribuciones de datos num√©ricos son histogramas y diagramas de caja , as√≠ que usemos la biblioteca **matplotlib** de Python para crear uno de cada uno de estos para la columna de **alquileres .**

```python
import pandas as pd
import matplotlib.pyplot as plt

# This ensures plots are displayed inline in the Jupyter notebook
%matplotlib inline

# Get the label column
label = bike_data['rentals']


# Create a figure for 2 subplots (2 rows, 1 column)
fig, ax = plt.subplots(2, 1, figsize = (9,12))

# Plot the histogram   
ax[0].hist(label, bins=100)
ax[0].set_ylabel('Frequency')

# Add lines for the mean, median, and mode
ax[0].axvline(label.mean(), color='magenta', linestyle='dashed', linewidth=2)
ax[0].axvline(label.median(), color='cyan', linestyle='dashed', linewidth=2)

# Plot the boxplot   
ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

# Add a title to the Figure
fig.suptitle('Rental Distribution')

# Show the figure
fig.show()
```

Las parcelas muestran que el n√∫mero de alquileres diarios oscila entre 0 y poco m√°s de 3.400. Sin embargo, el n√∫mero medio (y mediano) de alquileres diarios est√° m√°s cerca del extremo inferior de ese rango, con la mayor√≠a de los datos entre 0 y alrededor de 2200 alquileres. Los pocos valores por encima de este se muestran en el diagrama de caja como c√≠rculos peque√±os, lo que indica que son valores at√≠picos , en otras palabras, valores inusualmente altos o bajos m√°s all√° del rango t√≠pico de la mayor√≠a de los datos.

Podemos hacer el mismo tipo de exploraci√≥n visual de las caracter√≠sticas num√©ricas. Creemos un histograma para cada uno de estos.

```python
# Plot a histogram for each numeric feature
for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = bike_data[col]
    feature.hist(bins=100, ax = ax)
    ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
```

Las caracter√≠sticas num√©ricas parecen estar distribuidas de manera m√°s normal , con la media y la mediana m√°s cerca del centro del rango de valores, coincidiendo con los valores que ocurren con mayor frecuencia.

Nota : Las distribuciones no son realmente normales en el sentido estad√≠stico, lo que dar√≠a como resultado un histograma de "curva de campana" suave y sim√©trica con la media y la moda (el valor m√°s com√∫n) en el centro; pero generalmente indican que la mayor√≠a de las observaciones tienen un valor cercano al medio.
Hemos explorado la distribuci√≥n de los valores num√©ricos en el conjunto de datos, pero ¬øqu√© pasa con las caracter√≠sticas categ√≥ricas? Estos no son n√∫meros continuos en una escala, por lo que no podemos usar histogramas; pero podemos trazar un gr√°fico de barras que muestre el recuento de cada valor discreto para cada categor√≠a.

```python
import numpy as np

# plot a bar plot for each categorical feature count
categorical_features = ['season','mnth','holiday','weekday','workingday','weathersit', 'day']

for col in categorical_features:
    counts = bike_data[col].value_counts().sort_index()
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    counts.plot.bar(ax = ax, color='steelblue')
    ax.set_title(col + ' counts')
    ax.set_xlabel(col) 
    ax.set_ylabel("Frequency")
plt.show()
```

Muchas de las caracter√≠sticas categ√≥ricas muestran una distribuci√≥n m√°s o menos uniforme (lo que significa que hay aproximadamente el mismo n√∫mero de filas para cada categor√≠a). Las excepciones a esto incluyen:

* **feriados :** hay muchos menos d√≠as que son feriados que d√≠as que no lo son.
* **jornada laboral :** hay m√°s d√≠as laborables que no laborables.
* **weathersit :** La mayor√≠a de los d√≠as son de categor√≠a 1 (despejados), siendo la categor√≠a 2 (niebla y nubes) la siguiente m√°s com√∫n. Hay comparativamente pocos d√≠as de categor√≠a 3 (lluvia ligera o nieve) y ning√∫n d√≠a de categor√≠a 4 (lluvia intensa, granizo o niebla).
Ahora que sabemos algo sobre la distribuci√≥n de los datos en nuestras columnas, podemos comenzar a buscar relaciones entre las caracter√≠sticas y la etiqueta de **alquileres** que queremos poder predecir.

Para las caracter√≠sticas num√©ricas, podemos crear diagramas de dispersi√≥n que muestren la intersecci√≥n de los valores de caracter√≠sticas y etiquetas. Tambi√©n podemos calcular la estad√≠stica de correlaci√≥n para cuantificar la relaci√≥n aparente.

```python
for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = bike_data[col]
    label = bike_data['rentals']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs ' + col + '- correlation: ' + str(correlation))
plt.show()
```


Los resultados no son concluyentes, pero si observa de cerca los diagramas de dispersi√≥n de **temperatura** y **temperatura** , puede ver una tendencia diagonal vaga que muestra que los recuentos de alquileres m√°s altos tienden a coincidir con temperaturas m√°s altas; y un valor de correlaci√≥n de poco m√°s de 0,5 para ambas caracter√≠sticas respalda esta observaci√≥n. Por el contrario, las gr√°ficas de **zumbido** y **velocidad del viento** muestran una correlaci√≥n ligeramente negativa, lo que indica que hay menos alquileres en d√≠as con alta humedad o velocidad del viento.

Ahora comparemos las caracter√≠sticas categ√≥ricas con la etiqueta. Haremos esto mediante la creaci√≥n de diagramas de caja que muestren la distribuci√≥n de los recuentos de alquiler para cada categor√≠a.

```python
# plot a boxplot for the label by each categorical feature
for col in categorical_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    bike_data.boxplot(column = 'rentals', by = col, ax = ax)
    ax.set_title('Label by ' + col)
    ax.set_ylabel("Bike Rentals")
plt.show()
```


Las gr√°ficas muestran alguna variaci√≥n en la relaci√≥n entre algunos valores de categor√≠a y alquileres. Por ejemplo, existe una clara diferencia en la distribuci√≥n de los alquileres los fines de semana ( **d√≠a laborable** 0 o 6) y los de la semana laboral ( **d√≠a laborable** 1 a 5). Del mismo modo, existen diferencias notables entre las categor√≠as de **d√≠as festivos** y **d√≠as laborables** . # Entrenar un modelo de regresi√≥n
Ahora que hemos explorado los datos, es hora de usarlos para entrenar un modelo de regresi√≥n que use las caracter√≠sticas que hemos identificado como potencialmente predictivas para predecir la etiqueta de alquileres . Lo primero que debemos hacer es separar las caracter√≠sticas que queremos usar para entrenar el modelo de la etiqueta que queremos que prediga.

```python
# Separate features and labels
X, y = bike_data[['season','mnth', 'holiday','weekday','workingday','weathersit','temp', 'atemp', 'hum', 'windspeed']].values, bike_data['rentals'].values
print('Features:',X[:10], '\nLabels:', y[:10], sep='\n')
```


Despu√©s de separar el conjunto de datos, ahora tenemos numerosas matrices llamadas X que contienen las caracter√≠sticas e y que contienen las etiquetas.

Nos pudimos entrenar un modelo utilizando todos los datos; pero es una pr√°ctica com√∫n en el aprendizaje supervisado dividir los datos en dos subconjuntos; un conjunto (normalmente m√°s grande) con el que entrenar el modelo y un conjunto de "retenci√≥n" m√°s peque√±o con el que validar el modelo entrenado. Esto nos permite evaluar qu√© tan bien se desempe√±a el modelo cuando se usa con el conjunto de datos de validaci√≥n comparando las etiquetas predichas con las etiquetas conocidas. Es importante dividir los datos al azar (en lugar de decir, tomar el primer 70% de los datos para entrenar y guardar el resto para validaci√≥n). Esto ayuda a garantizar que los dos subconjuntos de datos sean estad√≠sticamente comparables (por lo que validamos el modelo con datos que tienen una distribuci√≥n estad√≠stica similar a los datos sobre los que se entren√≥).

Para dividir aleatoriamente los datos, usaremos la funci√≥n **train_test_split** en la biblioteca **scikit-learn** . Esta biblioteca es uno de los paquetes de aprendizaje autom√°tico m√°s utilizados para Python.

```python
from sklearn.model_selection import train_test_split

# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

Ahora tenemos los siguientes cuatro conjuntos de datos:

* **X_train :** los valores de caracter√≠sticas que usaremos para entrenar el modelo
* **y_train :** Las etiquetas correspondientes que usaremos para entrenar el modelo.
* **X_test :** los valores de caracter√≠sticas que usaremos para validar el modelo
* **y_test :** las etiquetas correspondientes que usaremos para validar el modelo
Ahora estamos listos para entrenar un modelo ajustando un algoritmo de regresi√≥n adecuado a los datos de entrenamiento. Usaremos un algoritmo de regresi√≥n lineal , un punto de partida com√∫n para la regresi√≥n que funciona tratando de encontrar una relaci√≥n lineal entre los valores X y la etiqueta y . El modelo resultante es una funci√≥n que define conceptualmente una l√≠nea en la que se cruzan todas las combinaciones posibles de valores X e y.

En **Scikit-Learn**, los algoritmos de entrenamiento est√°n encapsulados en estimadores , y en este caso usaremos el estimador LinearRegression para entrenar un modelo de regresi√≥n lineal.

```python
# Train the model
from sklearn.linear_model import LinearRegression

# Fit a linear regression model on the training set
model = LinearRegression().fit(X_train, y_train)
print (model)
```


# Evaluar el modelo entrenado
Ahora que hemos entrenado el modelo, podemos usarlo para predecir los recuentos de alquiler de las funciones que retuvimos en nuestro conjunto de datos de validaci√≥n. Luego, podemos comparar estas predicciones con los valores reales de la etiqueta para evaluar qu√© tan bien (¬°o no!) Est√° funcionando el modelo.

```python
import numpy as np

predictions = model.predict(X_test)
np.set_printoptions(suppress=True)
print('Predicted labels: ', np.round(predictions)[:10])
print('Actual labels   : ' ,y_test[:10])
```


Comparar cada predicci√≥n con su correspondiente valor real de "verdad fundamental" no es una forma muy eficiente de determinar qu√© tan bien est√° prediciendo el modelo. Veamos si podemos obtener una mejor indicaci√≥n visualizando un diagrama de dispersi√≥n que compare las predicciones con las etiquetas reales. Tambi√©n superpondremos una l√≠nea de tendencia para tener una idea general de qu√© tan bien se alinean las etiquetas predichas con las etiquetas verdaderas.

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```


Hay una tendencia diagonal definida, y las intersecciones de los valores predichos y reales generalmente siguen el camino de la l√≠nea de tendencia; pero hay una gran diferencia entre la funci√≥n ideal representada por la l√≠nea y los resultados. Esta varianza representa los residuos del modelo; en otras palabras, la diferencia entre la etiqueta predicha cuando el modelo aplica los coeficientes que aprendi√≥ durante el entrenamiento a los datos de validaci√≥n y el valor real de la etiqueta de validaci√≥n. Estos residuos, cuando se eval√∫an a partir de los datos de validaci√≥n, indican el nivel de error esperado cuando el modelo se utiliza con datos nuevos para los que se desconoce la etiqueta.

Puede cuantificar los residuos calculando una serie de m√©tricas de evaluaci√≥n de uso com√∫n. Nos centraremos en los siguientes tres:

* **Error cuadr√°tico medio (MSE) :** la media de las diferencias cuadr√°ticas entre los valores predichos y reales. Esto produce una m√©trica relativa en la que cuanto menor sea el valor, mejor ser√° el ajuste del modelo.
* **Error cuadr√°tico medio (RMSE) :** La ra√≠z cuadrada del MSE. Esto produce una m√©trica absoluta en la misma unidad que la etiqueta (en este caso, el n√∫mero de alquileres). Cuanto menor sea el valor, mejor ser√° el modelo (en un sentido simplista, representa el n√∫mero medio de alquileres en los que las predicciones son incorrectas).
* **Coeficiente de determinaci√≥n (generalmente conocido como R cuadrado o R2 ):** una m√©trica relativa en la que cuanto mayor sea el valor, mejor ser√° el ajuste del modelo. En esencia, esta m√©trica representa qu√© parte de la varianza entre los valores de etiqueta predichos y reales es capaz de explicar el modelo.
Nota : Puede obtener m√°s informaci√≥n sobre estas y otras m√©tricas para evaluar modelos de regresi√≥n en la documentaci√≥n de Scikit-Learn.
Usemos Scikit-Learn para calcular estas m√©tricas para nuestro modelo, en funci√≥n de las predicciones que gener√≥ para los datos de validaci√≥n.

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(y_test, predictions)
print("R2:", r2)
```


As√≠ que ahora hemos cuantificado la capacidad de nuestro modelo para predecir el n√∫mero de alquileres. Definitivamente tiene cierto poder predictivo, ¬°pero probablemente podamos hacerlo mejor!

# Regresi√≥n: experimentar con modelos adicionales
En el cuaderno anterior, usamos modelos de regresi√≥n simple para observar la relaci√≥n entre las caracter√≠sticas de un conjunto de datos de alquiler de bicicletas. En este cuaderno, experimentaremos con modelos m√°s complejos para mejorar nuestro rendimiento de regresi√≥n.

Comencemos cargando los datos de uso compartido de bicicletas como un Pandas DataFrame y viendo las primeras filas. Tambi√©n dividiremos nuestros datos en conjuntos de datos de prueba y entrenamiento.



```python
# Import modules we'll need for this notebook
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# load the training dataset
!wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/daily-bike-share.csv
bike_data = pd.read_csv('daily-bike-share.csv')
bike_data['day'] = pd.DatetimeIndex(bike_data['dteday']).day
numeric_features = ['temp', 'atemp', 'hum', 'windspeed']
categorical_features = ['season','mnth','holiday','weekday','workingday','weathersit', 'day']
bike_data[numeric_features + ['rentals']].describe()
print(bike_data.head())


# Separate features and labels
# After separating the dataset, we now have numpy arrays named **X** containing the features, and **y** containing the labels.
X, y = bike_data[['season','mnth', 'holiday','weekday','workingday','weathersit','temp', 'atemp', 'hum', 'windspeed']].values, bike_data['rentals'].values

# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

```python
from sklearn.linear_model import Lasso

# Fit a lasso model on the training set
model = Lasso().fit(X_train, y_train)
print (model, "\n")

# Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_text

# Train the model
model = DecisionTreeRegressor().fit(X_train, y_train)
print (model, "\n")

# Visualize the model tree
tree = export_text(model)
print(tree)
```

```python
# Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train the model
model = RandomForestRegressor().fit(X_train, y_train)
print (model, "\n")

# Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```
